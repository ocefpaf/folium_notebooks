```python
import json
import pandas
import folium


data = pandas.read_csv('data/US_Unemployment_Oct2012.csv')
mapa = folium.Map(location=[48, -102], zoom_start=3)
geo_path = 'data/us-states.json'

geo_str = json.dumps(json.load(open(geo_path, 'r')))

mapa.choropleth(geo_str=geo_str,
                data=data,
                columns=['State', 'Unemployment'],
                fill_color='YlGn',
                key_on='feature.id')


p = folium.RegularPolygonMarker(location=[45.5318, -122.6745],
                                color='red',
                                opacity=1,
                                weight=2,
                                fill_color='blue',
                                fill_opacity=1,
                                number_of_sides=8,
                                radius=10)

mapa.add_children(p)
mapa
```

```python
from folium.utilities import split_six

mapa = folium.Map(location=[48, -102], zoom_start=3)

mapa.choropleth(geo_path=geo_path,
                data=data,
                columns=['State', 'Unemployment'],
                fill_color='YlGn',
                key_on='feature.id',
                threshold_scale=split_six(data['Unemployment']))

mapa
```
